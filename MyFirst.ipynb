Exploratory Data Analysis

On the basis on GENDER

```python
ax=sns.countplot(x='Gender', data=df)
for bars in  ax.containers:
  ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data = sales_gen)
```

Conclusion On The Basis Of Gender- Most of the buyers  are female and even the amount purchased by females are greater than males

On the basis of Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')
```

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group', y='Amount', data=sales_age)
```

Conclusiom on the basis of Age Group-Most of the buyers are female and are of age group 26-35.The amount purchased by females and their count both are greater than males.

On the basis of State

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sales_state
```

```python
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

Conclusion on the basis of State-Most of the orders and maximum amount are from Uttar Pradesh,Karnataka and Maharashtra respectively

On the basis of Marital Status

```python
ax=sns.countplot(data=df, x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_status = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_status
```

```python
sales_status = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
sns.barplot(data=sales_status,x='Marital_Status', y='Amount', hue='Gender')
```

Conclusion on the basis of Marital Status-Most of the buyers are married females and their purchasing power is also greater than other categories.

On the basis of Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(data=df, x='Occupation')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_status=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data= sales_status, x='Occupation', y='Amount')
```

Conclusion on the basis of Occupation-Most of the buyers are from IT sector

```python
On the basis of  Product Category
```

```python
sns.set(rc={'figure.figsize': (20,5)})
ax=sns.countplot( data=df, x='Product_Category')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales_state=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sales_state
```

```python
sales_state=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category', y='Amount', data=sales_state)
```

Conclusion on the basis of Product Category-Most of the orders and maximum  amount spent are from Food,Clothing and Electronics.

**OVERALL CONCLUSION- Married Women of age 26-35 years from UP,Maharashtra and Karnataka those are from IT, HealthCare and Aviation industry are more likely to buy Food, Clothing and Electronics around Diwali**
